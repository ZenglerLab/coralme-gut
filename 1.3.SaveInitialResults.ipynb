```python
from build import parse_diet,constrain_diet
diet = pandas.read_csv("./diets/high_fiber_diet.txt",index_col=0,sep='\t',comment='#',header=None)
diet.columns = ["lb"]
diet_dct = parse_diet(diet)
```

```python
def get_objective(model):
    return next(r for r in model.reactions if r.objective_coefficient)
def get_biomass(model):
    r = get_objective(model)
    dct = {m.id:c for m,c in r.metabolites.items()}
    return pandas.DataFrame.from_dict({"coeff":dct})
def run(org):
    model = load_model(org)
    model.solution.to_frame().to_csv("./fluxes_reconstruction/me/{}.csv".format(org))
    constrain_diet(model.gem,diet_dct)
    model.gem.optimize().to_frame().to_csv("./fluxes_reconstruction/m/{}.csv".format(org))
    get_biomass(model.gem).to_csv("./biomass/{}.csv".format(org))
```

```python
for org in run_for:
    run(org)
```
