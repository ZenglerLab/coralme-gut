```python
from coralme.builder.helper_functions import flux_based_reactions
pandas.set_option('display.max_colwidth', None)
plt.rcParams['svg.fonttype'] = 'none'
```

```python
name = "bile"
fluxdir = "./cases/fluxes/{}".format(name)
if not os.path.isdir(fluxdir):
    os.makedirs(fluxdir)
```

```python
from knockouts import *
def run(org): 
    model = coralme.io.pickle.load_pickle_me_model("./me-models/Bifidobacterium_bifidum_BGN4/MEModel-BIO-Bifidobacterium_bifidum_BGN4-ME-TS.pkl")
    gem = model.gem
    rlist = []
    for r in gem.reactions:
        if "bile" in r.name:
            # print(r)
            rlist.append(r)
    melist = []
    for r in rlist:
        rxn = coralme.core.reaction.MetabolicReaction(r.id + "_ACYLTRANSFERASE")
        rxn.name = r.name + " (REVERSED)"
        model.add_reactions([rxn])
        rxn.add_metabolites({k:-v for k,v in r.metabolites.items()})
        melist.append(rxn)

    model.reactions.get_by_id("EX_cholate(e)").lower_bound = -1000
    
    index_dct = {r.id : model.reactions.index(r) for r in model.reactions}
    met_index_dct = {m.id : model.metabolites.index(m) for m in model.metabolites}
    max_mu = model.solution.objective_value*2
    me_nlp = get_nlp(model)
    del model
    sol = optimize(index_dct,met_index_dct,me_nlp,max_mu=max_mu)
    sol.to_frame().to_csv("cases/fluxes/{}/{}.csv".format(name,org))
```

```python
organisms = set(pandas.read_csv("biomass_constrained.txt",index_col=0,header=None).index.to_list())
bile_models = set([i.split(".json")[0] for i in pandas.read_csv("./bile_salt_models.txt",index_col=0).index.to_list()])
```

```python
run_for = organisms & bile_models
len(run_for)
```

```python
NP = min([10,len(run_for)])
pool = mp.Pool(NP,maxtasksperchild=1)
pbar = tqdm(total=len(run_for),position=0,leave=True)
pbar.set_description('Building ({} threads)'.format(NP))
def collect_result(result):
    pbar.update(1)

for org in run_for:
    args = ([org])
    pool.apply_async(run,args, callback=collect_result)
pool.close()
pool.join()
```
