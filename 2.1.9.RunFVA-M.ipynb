```python
from diets import load_model,constrain_exchanges,get_bounds
from overflow import get_overflow
import coralme
import sympy
mu = sympy.Symbol("mu", positive = True)
def run(org):
    gem = coralme.io.pickle.load_pickle_me_model("./me-models/{}/MEModel-BIO-{}-ME-TS.pkl".format(org,org)).gem
    model = coralme.core.model.MEModel(gem.id)
    rlist = []

    for m in gem.metabolites:
        met = coralme.core.metabolites.Metabolite(m.id)
    
    for r in gem.reactions:
        rxn = coralme.core.reactions.MEReaction(r.id)
        model.add_reactions([rxn])
        for m,c in r.metabolites.items():
            rxn.add_metabolites(
                {m.id : c}
            )
    rxn = model.get("biomass_dilution")
    rxn.add_metabolites({
        "biomass_c" : -1,
        "biomass" : 0
    },combine=False)
    rxn.bounds = (mu,mu)

    model.optimize()
    
    exchanges = [r.id for r in model.reactions.query("^EX_")]
    res = model.fva(
		exchanges, 1.0, mu_fixed = None,
		max_mu = 2.8100561374051836, min_mu = 0., maxIter = 100, lambdify = True,
		tolerance = 1e-6, precision = 'quad', verbose = True)
    res.to_csv("./cases/fva/{}/{}.csv".format(name,org))
```

```python
done = set(i.split(".csv")[0] for i in os.listdir("./cases/fva/{}/".format(name)) if "csv" in i)
run_for = organisms - done
len(run_for)
```

```python
run(list(run_for)[0])
```

```python
NP = min([10,len(run_for)])
pool = mp.Pool(NP,maxtasksperchild=1)
pbar = tqdm(total=len(run_for),position=0,leave=True)
pbar.set_description('Building ({} threads)'.format(NP))
def collect_result(result):
    pbar.update(1)

for org in run_for:
    args = ([org])
    pool.apply_async(run,args, callback=collect_result)
pool.close()
pool.join()
```
